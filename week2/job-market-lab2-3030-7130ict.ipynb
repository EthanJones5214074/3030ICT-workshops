```python
#new dataframe for location data
df = pd.DataFrame(data)

# Using groupby() and count()
locationDF = df['Location'].value_counts()
#locationDF.sort_values(by=)
locationDF.head()
```

```python
ax = locationDF.plot.barh(x='col2', y='col1')
```
